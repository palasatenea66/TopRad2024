```python
# =======================
# Sección 1: Introducción a Pandas y Matplotlib
# =======================
# Pandas es una biblioteca de Python para el análisis de datos, mientras que Matplotlib se usa para graficar datos.
# Con Pandas podemos estructurar y manipular datos, y luego visualizarlos con Matplotlib.

# Importamos las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Activar la visualización en línea para Jupyter
%matplotlib inline

print("¡Bibliotecas listas para usarse!")
```

```python
# =======================
# Sección 2: Creación de un DataFrame para Graficar
# =======================
# Creamos un DataFrame de ejemplo con datos para graficar

# Datos de ventas por producto en diferentes ciudades
data = {
    "Producto": ["A", "B", "C", "D"],
    "Ventas": [300, 500, 700, 200],
    "Costos": [200, 300, 400, 150]
}
df = pd.DataFrame(data)
print("DataFrame de ventas y costos:")
df

# Ejercicio: Crea un DataFrame con tus propios datos, como temperaturas de ciudades o calificaciones.
```

```python
# =======================
# Sección 3: Graficar una Columna (Gráfico de Barras)
# =======================
# Vamos a graficar la columna 'Ventas' en un gráfico de barras.

plt.figure(figsize=(8, 6))  # Tamaño del gráfico
plt.bar(df["Producto"], df["Ventas"], color='skyblue')
plt.xlabel("Producto")
plt.ylabel("Ventas")
plt.title("Ventas por Producto")
plt.show()

# Ejercicio: Cambia el gráfico para que grafique la columna 'Costos' en lugar de 'Ventas'.
```

```python
# =======================
# Sección 4: Graficar Varias Columnas en el Mismo Gráfico
# =======================
# Podemos graficar múltiples columnas en el mismo gráfico para comparar datos.

plt.figure(figsize=(8, 6))
plt.bar(df["Producto"], df["Ventas"], color='skyblue', label="Ventas")
plt.bar(df["Producto"], df["Costos"], color='salmon', label="Costos", alpha=0.7)
plt.xlabel("Producto")
plt.ylabel("Monto")
plt.title("Ventas y Costos por Producto")
plt.legend()  # Mostrar leyenda
plt.show()

# Ejercicio: Crea un gráfico de barras que compare otras dos columnas en tu propio DataFrame.
```

```python
# =======================
# Sección 5: Gráfico de Líneas
# =======================
# Los gráficos de líneas son útiles para mostrar tendencias en los datos a lo largo del tiempo o de otro eje.

# Creamos datos de ejemplo de ventas semanales
data_semanal = {
    "Semana": [1, 2, 3, 4, 5],
    "Ventas": [150, 200, 300, 250, 400],
    "Costos": [100, 150, 200, 180, 210]
}
df_semanal = pd.DataFrame(data_semanal)

# Graficamos las ventas y los costos en el tiempo (por semana)
plt.figure(figsize=(10, 6))
plt.plot(df_semanal["Semana"], df_semanal["Ventas"], marker='o', color='blue', label="Ventas")
plt.plot(df_semanal["Semana"], df_semanal["Costos"], marker='s', color='red', label="Costos")
plt.xlabel("Semana")
plt.ylabel("Monto")
plt.title("Ventas y Costos Semanales")
plt.legend()
plt.grid()
plt.show()

# Ejercicio: Agrega otra columna de datos (como ganancias) y agrégala al gráfico.
```

```python
# =======================
# Sección 6: Gráfico Circular (Pie Chart)
# =======================
# Un gráfico circular es útil para mostrar proporciones.

# Creamos datos de ejemplo para un gráfico circular
data_pie = {
    "Categoría": ["Categoría A", "Categoría B", "Categoría C", "Categoría D"],
    "Porcentaje": [25, 35, 20, 20]
}
df_pie = pd.DataFrame(data_pie)

# Graficamos el gráfico circular
plt.figure(figsize=(7, 7))
plt.pie(df_pie["Porcentaje"], labels=df_pie["Categoría"], autopct="%1.1f%%", startangle=90)
plt.title("Distribución de Categorías")
plt.show()

# Ejercicio: Crea tu propio gráfico circular con diferentes datos.
```

```python
# =======================
# Sección 7: Histograma
# =======================
# Los histogramas son útiles para mostrar la distribución de un conjunto de datos.

# Creamos un conjunto de datos aleatorio para el histograma
import numpy as np
data_hist = {
    "Edad": np.random.randint(18, 60, size=100)  # 100 edades aleatorias entre 18 y 60
}
df_hist = pd.DataFrame(data_hist)

# Graficamos el histograma de la columna 'Edad'
plt.figure(figsize=(8, 6))
plt.hist(df_hist["Edad"], bins=10, color='purple', edgecolor='black')
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de Edad")
plt.show()

# Ejercicio: Cambia el número de bins (intervalos) para ver cómo cambia el histograma.
```

```python
# =======================
# Sección 8: Gráficos de Dispersión (Scatter Plot)
# =======================
# Los gráficos de dispersión son útiles para ver la relación entre dos variables numéricas.

# Creamos datos de ejemplo de precios y ventas
data_dispersion = {
    "Precio": np.random.randint(50, 300, size=50),
    "Ventas": np.random.randint(200, 1500, size=50)
}
df_dispersion = pd.DataFrame(data_dispersion)

# Graficamos la relación entre precio y ventas
plt.figure(figsize=(8, 6))
plt.scatter(df_dispersion["Precio"], df_dispersion["Ventas"], color='darkgreen')
plt.xlabel("Precio")
plt.ylabel("Ventas")
plt.title("Relación entre Precio y Ventas")
plt.show()

# Ejercicio: Cambia los datos de ejemplo para mostrar otra relación entre variables numéricas.
```

```python
# =======================
# Sección 9: Personalización de Gráficos
# =======================
# Podemos personalizar los gráficos con diferentes colores, tamaños, estilos y más.

# Gráfico de barras personalizado
plt.figure(figsize=(10, 6))
plt.bar(df["Producto"], df["Ventas"], color='royalblue', edgecolor='black', linewidth=1.5)
plt.xlabel("Producto", fontsize=12)
plt.ylabel("Ventas", fontsize=12)
plt.title("Ventas por Producto (Personalizado)", fontsize=14, fontweight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Ejercicio: Crea un gráfico personalizado con diferentes colores, etiquetas y estilos.
```

```python
# =======================
# Sección 10: Guardar Gráficos
# =======================
# Podemos guardar gráficos generados por Matplotlib en archivos de imagen.

# Creamos un gráfico y lo guardamos como archivo PNG
plt.figure(figsize=(8, 6))
plt.plot(df_semanal["Semana"], df_semanal["Ventas"], marker='o', color='blue', label="Ventas")
plt.plot(df_semanal["Semana"], df_semanal["Costos"], marker='s', color='red', label="Costos")
plt.xlabel("Semana")
plt.ylabel("Monto")
plt.title("Ventas y Costos Semanales")
plt.legend()
plt.grid()

# Guardar el gráfico en un archivo
plt.savefig("grafico_semanal.png")
print("Gráfico guardado como 'grafico_semanal.png'")

# Ejercicio: Guarda otro gráfico de tu elección en formato PNG o JPG.
```
